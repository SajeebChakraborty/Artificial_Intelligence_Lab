```python
#import necessaries
import tensorflow as tf
import numpy as np
```

```python
# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, input_shape=[1], activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
```

```python
X = np.random.uniform(-10, 10, (10000, 1))
y = 3 * X**3 + 7 * X**2 - 12 * X + 2
```

```python
# Train the model
model.fit(X, y, epochs=300, batch_size=32)
```

```python
#input data
inData = [0,1,-1,-2,2,3,-3]
out = model.predict(inData)

#calculate length of data
num = len(inData)

for i in range(num):

  #orginal calculated y for data
  orginal_out = 3 * inData[i] * inData[i] * inData[i] + 7 * inData[i] * inData[i] - 12 * inData[i] + 2

  #compare predicted value of y with original for input data
  print(f"x : {inData[i]} =>  original y : {orginal_out} => pre y : {float(out[i])}")
```
